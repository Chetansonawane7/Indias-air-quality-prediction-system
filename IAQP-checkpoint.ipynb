```python
import pandas as pd
import numpy as np
from prophet import Prophet
```

```python
air_quality_data = pd.read_csv('air-quality-india.csv')
```

```python
print(air_quality_data.head(5))
```

```python
pollution = air_quality_data.drop(['Year', 'Month', 'Day', 'Hour'],axis=1)
print(pollution)
```

```python
pollution.info()
```

```python
pollution.isnull().sum
```

```python
pollution = pd.DataFrame(pollution)
```

```python
print(pollution)
```

```python
pollution.isnull().sum
```

```python
print(pollution.tail(5))
```

```python
pollution['ds'] = pollution['Timestamp']
```

```python
print(pollution.head(2))
```

```python
pollution = pollution.drop(['ds'],axis=1)
print(pollution)
```

```python
pollution.rename(columns = {'Timestamp':'ds'}, inplace = True) 
```

```python
print(pollution.head(2))
```

```python
pollution.rename(columns = {'PM2.5':'y'}, inplace = True) 
print(pollution.head(2))
```

```python
model = Prophet()
model.fit(pollution)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```
